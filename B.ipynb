```python
import numpy as np 
import pandas as pd 
```

```python
X = np.array([[5,10,1],[7,3,1],[100,5,1]])
```

```python
Y = np.array([[1,0,0],[0,1,0],[0,0,1]])
```

```python
w_1 = np.array([[0.2,0.4,1],[0.15,0.75,1],[0.3,0.01,1]])
```

```python
w_2 = np.array([[0.9,0.25,0.12],[0.99,0.01,0.5]])
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def sigmoid_der(x):
    return sigmoid(x)*(1-sigmoid(x))
```

```python
for back in range(10):
    
    in_o = np.dot(X, w_1)

    out_o = sigmoid(in_o)

    error = out_o - Y

    x = error.sum()
    print(x)

    derror_douto = error
    douto_dino = sigmoid_der(out_o)

    deriv = derror_douto * douto_dino

    deriv_final = np.dot(X.T, deriv)

    w_1 -= deriv_final

```
